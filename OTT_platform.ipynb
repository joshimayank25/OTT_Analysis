```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movie['genres']= movie['genres'].apply(convert)
```

```python
import ast

def convert1(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count=count+1
        else:
            break
    return L
```

```python
movie['cast'] = movie['cast'].apply(convert1)
```

```python
import ast

def fetch(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movie['crew'] = movie['crew'].apply(fetch)
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" "," ") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" "," ") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" "," ") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" "," ") for i in x])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(new_movie['tags']).toarray()
```

```python
vector
```

```python
vector[0]
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('loving')
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_movie['tags'] = new_movie['tags'].apply(stem)
```

```python
new_movie
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[2]
```

```python
similarity[0]
```

```python
sorted(similarity[0], reverse=True)
```

```python
list(enumerate(similarity[0]))
```

```python
def recommend(movie):
    movie_index = new_movie[new_movie['title'] == movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_movie.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
